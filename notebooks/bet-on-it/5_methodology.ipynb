# Post infall

I will want to plot full orb as the dashed, and masked orb as solid
Also, want to use the arg of merger to make a triangle at time of merger!

```python
fig, ax = plt.subplots(3,1, figsize=(8,12), sharex=True, 
                       gridspec_kw={'width_ratios': [1], 
                                    'height_ratios':[1,1,1],
                                    'wspace': 0.10,
                                    "hspace":0.0})

# note: switched these to be subfind ids instead of indices
mergers = [446120, 277085, 282026, 309714, 397340]
likely_subind = [320584, 447958, 464542] # 92, 112
unbound_subind = [419008, 412090, 383035, 431867, 360801] # 94
cmap_name = ["gist_heat","winter","summer"]
infall = np.array(np.triu(np.zeros((100,100))+1),dtype='bool')

for plotnum, subfindids in enumerate([mergers,likely_subind,unbound_subind]):
    # first, plot vertical line at z=1.5
    lbt_z1p5 = convert['Lookback Time'][convert['Snapshot']==40]
    ax[plotnum].axvline(lbt_z1p5, color='black', lw=0.5, alpha=0.5)
    
    # find corresponding indices of full table
    inds = [np.where(orbit['SubfindID1']==idd)[0][0] for idd in subfindids]

    # initialize colormaps 
    colorval = np.linspace(0.2,0.8,len(inds))
    cmap = plt.get_cmap(cmap_name[plotnum])

    for en, ind in enumerate(inds):
        # for mergers, want to only select post merger and pre infall
        linecolor = cmap(colorval[en])
        fullorb = orbit['Separations'][ind].copy()
        post_infall = infall[orbit['InfallSnapshot']][ind]

        if plotnum == 0:    # mergers only
            # mask out part of orbit after merger (keeping only single sep=0 element )
            # note: only want to mask out merger if the pair DOES merge aka panel 1 only
            arg_of_merger = np.argwhere(fullorb==0)[0][0]
            merger_snap = snaps[arg_of_merger]
            next_arg = arg_of_merger + 1
            fullorb[next_arg:] = np.nan
            
            # next, mask out things before infall 
            maskedorb = np.where(post_infall, fullorb, np.nan)
            
            ax[plotnum].plot(snaps, fullorb, color=linecolor, linestyle='dashed', lw=1)  
            ax[plotnum].plot(snaps, maskedorb, color=linecolor, linestyle='solid', lw=2)  
            
            ax[plotnum].scatter(merger_snap, 0, color=linecolor, marker="^")

        else:    # there aren't mergers in these, so only need to select post-infall
            maskedorb = np.where(post_infall, fullorb, np.nan)

            ax[plotnum].plot(snaps, fullorb, color=linecolor, linestyle='dashed', lw=1)  
            ax[plotnum].plot(snaps, maskedorb, color=linecolor, linestyle='solid', lw=2)  
    
fig.supylabel("Separation [kpc]",fontsize=20)

xval = 0
ax[0].text(xval,320,"Mergers",ha='right',va='center',fontweight='medium')
ax[1].text(xval,440,"Potential future mergers",ha='right',va='center',fontweight='medium')
ax[2].text(xval+1.5,2450,"Fly-bys",ha='right',fontweight='medium')

ax[0].plot([0],[0],linestyle='dashed', lw=1,color='black',alpha=1,label='False')
ax[0].plot([0],[0],lw=3,color='black',alpha=1,label='True')
ax[0].legend(loc='center right',title="Post-Infall:",fontsize=14,title_fontsize=14,frameon=False)

plt.xlim(np.max(convert['Lookback Time']+0.5),-0.5)
plt.xlabel("Lookback time [Gyr]",fontsize=20)
plt.savefig("../../plots/bet-on-it/5_exampleorbits.png",bbox_inches='tight')
plt.show()
```
